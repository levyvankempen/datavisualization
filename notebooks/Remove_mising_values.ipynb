```python
import pandas as pd
```

```python
player_defense = pd.read_csv("../player_data/player_defense" + ".csv")
player_gca = pd.read_csv("../player_data/player_gca" + ".csv")
player_keepers = pd.read_csv("../player_data/player_keepers" + ".csv")
player_keepersadv = pd.read_csv("../player_data/player_keepersadv" + ".csv")
player_misc = pd.read_csv("../player_data/player_misc" + ".csv")
player_passing = pd.read_csv("../player_data/player_passing" + ".csv")
player_passing_types = pd.read_csv("../player_data/player_passing_types" + ".csv")
player_playingtime = pd.read_csv("../player_data/player_playingtime" + ".csv")
player_possession = pd.read_csv("../player_data/player_possession" + ".csv")
player_shooting = pd.read_csv("../player_data/player_shooting" + ".csv")
player_stats = pd.read_csv("../player_data/player_stats" + ".csv")
```

```python
list_of_names = [player_defense, player_gca, player_keepers, player_keepersadv, 
                 player_misc, player_passing, player_passing_types, player_playingtime,
                player_possession, player_shooting, player_stats]
```

```python
def remove_players_with_few_minutes(list_of_names):
    for i in list_of_names:
        i.drop(i[i['minutes_90s'] < 1.0].index, inplace=True)
```

```python
remove_players_with_few_minutes(list_of_names)
```

```python
def check_missing_values(player_dataset):
    print(player_dataset.isnull().sum())
```

```python
check_missing_values(player_defense)
```

```python
player_defense
```

```python
def show_missing_values(player_dataset, column):
    return player_dataset[player_dataset[column].isnull()]
```

```python
show_missing_values(player_defense, 'tackles')
```

```python
def set_missing_values_to_zero(player_dataset, column):
    player_dataset[column] = player_dataset[column].fillna(0)
```

```python
set_missing_values_to_zero(player_defense, 'dribble_tackles_pct')
```

```python
def remove_other_missing_values(player_dataset):
    player_dataset.dropna(inplace=True)
```

```python
remove_other_missing_values(player_defense)
```
